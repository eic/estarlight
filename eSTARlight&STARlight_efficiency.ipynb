```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Read the CSV file into a pandas DataFrame
dfJL = pd.read_csv('1000weight_eSTAR.csv', delimiter = ' ')
dfJL
```

```python
dfJL1 = dfJL["_weight"]*1
```

```python
plt.rc('font',size=18)
plt.rcParams['figure.figsize'] = [12, 8]
ax1 = plt.subplot()
plt.hist(dfJL1, bins=np.logspace(start=np.log10(10**(-9)), stop=np.log10(1), num=35))
plt.gca().set_xscale("log")
ax1.minorticks_on()
ax1.tick_params('both', length=15, width=2, which='major')
ax1.tick_params('both', length=7, width=1, which='minor')
plt.show()
```

```python
min(dfJL1), max(dfJL1)
```

```python
# Calculate the average of the entire DataFrame
average = np.mean(dfJL1)

# Print the average
print("Average of the data:", average)
```

```python
Efficiency = average/max(dfJL1)

# Print the efficiency
print("Effiency of weight distribution(1000) is {:.2%}".format(Efficiency))
```
